# Use a pretrained Transformer to predict SMILES

Load model: ncfrey/ChemGPT-1.2B - trained for predicting smiles

**Note: The code below will download an entire pretrained model and is quite large in size. This is intended just for demonstrative purposes, not recommended to perform on your local PC.**

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "ncfrey/ChemGPT-1.2B"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

model.eval()
```

Generate smiles:

```python
def generate_smiles(prompt="C", max_length=50):
    inputs = tokenizer(prompt, return_tensors="pt")

    outputs = model.generate(
        **inputs,
        max_length=max_length,
        do_sample=True,
        temperature=0.9,
        top_k=50,
        top_p=0.95
    )

    smiles = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return smiles
```
